### Stripped from Antonia's Google Sheet
Later we will bring this in from the raw data but I was on the hustle

```python
# organizations
s = """Baylor University
Boise State University
Boston University
Colorado School of Mines
Columbia University
Iowa State University
Joint Global Change Research Institute
Lawrence Berkeley National Laboratory
Lehigh University
Los Alamos National Laboratory
Massachusetts Institute of Technology
NASA Goddard Institute for Space Studies
National Renewable Energy Laboratory - Golden Campus
Oak Ridge National Laboratory
Penn State University
Pacific Northwest National Laboratory
Purdue University
Sandia National Laboratories, 4100 National Parks Hwy, Carlsbad, NM 88220
Stanford University
Tufts University
University of California Irvine
University of Denver
University of Houston
University of Illinois
University of New Hampshire
University of Texas at Austin"""

locations = s.split('\n')
```

```python
# number of particpants per organization
count = """1
3
2
1
1
1
8
3
1
1
7
1
5
3
6
26
4
2
1
1
1
1
1
1
1
1"""

ct_list = [int(i) for i in count.split('\n')]
```

```python
# connect to the Google Maps client
gmaps = googlemaps.Client(key=api_key)
```

```python
# create a dictionary of geocoded results to organization; we sleep to make it not time out from their server
d = {}

for i in locations:
    
    print(i)
    
    # Geocoding an address
    geocode_result = gmaps.geocode(i)
    
    d[i] = geocode_result
    
    time.sleep(2)
```

```python
# extract the lat long stuff from the geocoded result into it's own dict
dx = {"site": [], "latitude": [], "longitude": [], "participants": ct_list}

for i in d.keys():
    
    print(i)
    
    dx["site"].append(i)

    try:
        latitude = d[i][0]["geometry"]["location"]["lat"]
        longitude = d[i][0]["geometry"]["location"]["lng"]

        dx["latitude"].append(latitude)
        dx["longitude"].append(longitude)
        
    except IndexError:
        print(d[i])
```

```python
# to a data frame and then save it locally
df = pd.DataFrame(dx)

df.to_csv('/Users/d3y010/Desktop/cop.csv', index=False)
```

### Just fooling about here, fudging some fake interactions between orgs

```python
d_orgs = {}

choices = [1, 2, 3, 7]

orgs = df.site.values

n_orgs = orgs.shape[0]

for i in orgs:
    
    # get a random selection for the number of connections for testing
    n_choices = np.random.choice(choices)
    
    d_orgs[i] = []
    
    while n_choices > 0:

        selection = np.random.choice(orgs)

        if selection not in d_orgs[i]:
            d_orgs[i].append(selection)
            n_choices -= 1
        
    
x_dict = {"origin": [], "destination": []}

for i in d_orgs.keys():
    
    for j in d_orgs[i]:
        
        x_dict["origin"].append(i)
        x_dict["destination"].append(j)
        
xdf = pd.DataFrame(x_dict)

xdf.to_csv('/Users/d3y010/Desktop/cop_collab.csv', index=False)

xdf     
```
